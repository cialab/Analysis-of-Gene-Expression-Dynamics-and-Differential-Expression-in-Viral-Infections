```python

def get_sorted_genes(df, genes):
    sorted_genes_df = df[df['Gene Symbol'].isin(genes)].sort_values(by='MAS_rank', ascending=True)
    return sorted_genes_df['Gene Symbol'].tolist()

def get_sorted_overlap_genes(df1, df2, genes_overlap):
    combined_df = pd.concat([df1[df1['Gene Symbol'].isin(genes_overlap)], df2[df2['Gene Symbol'].isin(genes_overlap)]])
    combined_df['combined_MAS_rank'] = combined_df.groupby('Gene Symbol')['MAS_rank'].transform('sum')
    sorted_combined_df = combined_df.drop_duplicates('Gene Symbol').sort_values('combined_MAS_rank')
    return sorted_combined_df['Gene Symbol'].tolist()

def get_sorted_all_overlap_genes(df1, df2, df3, genes_all_overlap):
    combined_df = pd.concat([df1[df1['Gene Symbol'].isin(genes_all_overlap)], df2[df2['Gene Symbol'].isin(genes_all_overlap)], df3[df3['Gene Symbol'].isin(genes_all_overlap)]])
    combined_df['combined_MAS_rank'] = combined_df.groupby('Gene Symbol')['MAS_rank'].transform('sum')
    sorted_combined_df = combined_df.drop_duplicates('Gene Symbol').sort_values('combined_MAS_rank')
    return sorted_combined_df['Gene Symbol'].tolist()
```

```python
import pandas as pd

def find_min_max_mas_gene2(df1, df2, genes_all_overlap):
    # Filter dataframes for the genes of interest
    df1_filtered = df1[df1['Gene Symbol'].isin(genes_all_overlap)][['Gene Symbol', 'MAS_rank']]
    df2_filtered = df2[df2['Gene Symbol'].isin(genes_all_overlap)][['Gene Symbol', 'MAS_rank']]
    
    
    # Combine the filtered dataframes
    combined_df = pd.concat([df1_filtered, df2_filtered])
    
    # Find the maximum MAS_rank for each gene across the dataframes
    combined_df['max_MAS_rank'] = combined_df.groupby('Gene Symbol')['MAS_rank'].transform('max')
    
    # Drop duplicates to ensure each gene is represented once
    combined_df = combined_df.drop_duplicates('Gene Symbol')
    
    # Sort by 'max_MAS_rank' to get the gene with minimum of the maximum MAS ranks
    combined_df = combined_df.sort_values('max_MAS_rank')
    
    # Return the gene with the minimum of the maximum MAS ranks
    min_max_mas_gene = combined_df.iloc[0]['Gene Symbol']
    return min_max_mas_gene

```
